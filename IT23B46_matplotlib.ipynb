```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,marker="o")
```

```python
plt.plot(x,y,marker="*",ms="20")
#To create marker size
```

```python
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y")
#To give border color and fill color
#mec=marker end color
#mfc=marker fill color
```

```python
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashed")
#to give line style dashed
```

```python
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot")
#to give dashdot linestyle
```

```python
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="10")
#to give line width
```

```python
plt.title("Demo chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="5",color="g")
#To give labels to the axis and chart name and color to the axis
```

```python
x1=np.array([2,7,9,13,16])
y1=np.array([3,6,7,9,14])
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="5",color="g")
plt.plot(x1,y1,marker="o",ms="20",mec="g",mfc="b",linestyle="dashed",linewidth="5",color="r")
#creating multiple graphs
```

```python
font={'color':'green','size':8}
plt.title("Demo chart",fontdict=font,loc='right')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="5",color="g")
plt.plot(x1,y1,marker="o",ms="20",mec="g",mfc="b",linestyle="dashed",linewidth="5",color="r")
#to give decoration to title i.e fontsize and colors and location
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="5",color="g")
plt.subplot(1,2,2)
plt.plot(x1,y1,marker="o",ms="20",mec="g",mfc="b",linestyle="dashed",linewidth="5",color="r")
#Use subplot to divide graph into rows and columns
```

```python
plt.subplot(1,2,1)
plt.title("p1")
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="5",color="g")
plt.subplot(1,2,2)
plt.title("p2")
plt.plot(x1,y1,marker="o",ms="20",mec="g",mfc="b",linestyle="dashed",linewidth="5",color="r")
plt.suptitle("Supertitle")
#To give name to subplots and super plot
```

```python
plt.subplot(2,1,1)
plt.title("p1")
plt.plot(x,y,marker="*",ms="20",mec="r",mfc="y",linestyle="dashdot",linewidth="5",color="g")
plt.subplot(2,1,2)
plt.title("p2")
plt.plot(x1,y1,marker="o",ms="20",mec="g",mfc="b",linestyle="dashed",linewidth="5",color="r")
plt.suptitle("Supertitle")
#To change orientation of subplots according to rows and columns
```

```python
x=np.array([3,4,7,9])
y=np.array([3,6,4,5])
plt.plot(x,y)
```

```python
x1=np.array([7,4,5,8])
y1=np.array([2,6,7,3])
plt.plot(x1,y1)
plt.plot(x,y)
```

```python
plt.scatter(x,y,color='y',s=50,marker='*') # s=size,alpha=brightness
plt.scatter(x1,y1,s=40,alpha=0.5)
plt.show()
```

```python
plt.bar(x1,y1)
plt.show()
```

```python
plt.barh(x1,y1)
plt.show()
```

```python
plt.bar(x1,y1,color='y',width=0.5)
plt.show()
```

```python
plt.barh(x1,y1,color='r',height=0.5)
plt.show()
```

```python
plt.hist(x,color='b') #hist - generates histogram of column
```

```python
track_stud=np.array([20,40,50,10,15])
track_name=np.array(["web","data","Mobile","Cybersecurity","Iot"])
plt.pie(track_stud)
plt.show()
```

```python
track_stud=np.array([20,40,50,10,15])
track_name=np.array(["web","data","Mobile","Cybersecurity","Iot"])
plt.pie(track_stud,labels=track_name)# gives label to chart
plt.show()
```

```python
track_stud=np.array([20,40,50,10,15])
track_name=np.array(["web","data","Mobile","Cybersecurity","Iot"])
ex=[0.2,0,0,0.5,0]
c=['y',"darkred",'pink','m','b']
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True,colors=c)
plt.legend(title="TrackSelection")
plt.show()
```
